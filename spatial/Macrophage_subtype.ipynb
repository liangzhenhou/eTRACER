```r
T1_spatial <- readRDS('/syn1/liangzhen/jinhua_jilab_project/result/SPATIAL/T1/bin50_seurat.rds')
DefaultAssay(T1_spatial) <- 'RNA'
T1_spatial <- NormalizeData(T1_spatial)
RCTD <- readRDS('/syn1/liangzhen/jinhua_jilab_project/result/SPATIAL/T1/RCTD_results_doublet.rds')
T1_spatial <- AddMetaData(T1_spatial, metadata = RCTD@results$results_df)
T1_spatial@meta.data[T1_spatial@meta.data$spot_class != 'reject',]
```

```r
T2_spatial <- readRDS('/syn1/liangzhen/jinhua_jilab_project/result/SPATIAL/T2/bin50_seurat.rds')
DefaultAssay(T2_spatial) <- 'RNA'
T2_spatial <- NormalizeData(T2_spatial)
RCTD <- readRDS('/syn1/liangzhen/jinhua_jilab_project/result/SPATIAL/T2/RCTD_results_doublet.rds')
T2_spatial <- AddMetaData(T2_spatial, metadata = RCTD@results$results_df)
T2_spatial@meta.data[T2_spatial@meta.data$spot_class != 'reject',]
```

```r
T3_spatial <- readRDS('/syn1/liangzhen/jinhua_jilab_project/result/SPATIAL/T3a/bin50_seurat.rds')
DefaultAssay(T3_spatial) <- 'RNA'
T3_spatial <- NormalizeData(T3_spatial)
RCTD <- readRDS('/syn1/liangzhen/jinhua_jilab_project/result/SPATIAL/T3a/RCTD_results_doublet.rds')
T3_spatial <- AddMetaData(T3_spatial, metadata = RCTD@results$results_df)
T3_spatial@meta.data[T3_spatial@meta.data$spot_class != 'reject',]
```

```r
spatial <- merge(T1_spatial,y=list(T2_spatial,T3_spatial),add.cell.ids = c('T1','T2','T3'))
```

```r
spatial[['SCT']] <- NULL
```

```r
spatial@meta.data$time <- substr(rownames(spatial@meta.data),1,2)
```

```r
spatial@meta.data$time_cluster <- paste0(spatial@meta.data$time,'_',spatial@meta.data$first_type)
```

```r
spatial_macrophage <- subset(spatial,first_type == 'Macrophage' & spot_class != 'reject')
spatial_macrophage
```

```r
spatial_macrophage <- SCTransform(spatial_macrophage, vars.to.regress = "percent.mt", verbose = FALSE)
spatial_macrophage <- RunPCA(spatial_macrophage, verbose = FALSE)
spatial_macrophage <- RunUMAP(spatial_macrophage, dims = 1:30, verbose = FALSE)
```

```r
spatial_macrophage <- FindNeighbors(spatial_macrophage, dims = 1:30, reduction = "pca")
spatial_macrophage <- FindClusters(spatial_macrophage, resolution = 0.2)
```

```r
FeaturePlot(spatial_macrophage,features = 'Csf1r')
```

```r
options(repr.plot.width = 6, repr.plot.height = 6) 
DimPlot(spatial_macrophage, reduction = "umap",group.by = 'seurat_clusters',raster=FALSE,label = T,label.size = 8,cols =c('#A5B55D','#FD763F','#2472A3','#EECA40'))
```

```r
spatial_macrophage@meta.data$celltype <- 'None'
spatial_macrophage@meta.data[spatial_macrophage@meta.data$seurat_clusters == 0,'celltype'] <- 'Macro-Apoe'
spatial_macrophage@meta.data[spatial_macrophage@meta.data$seurat_clusters == 1,'celltype'] <- 'Macro-Arg1'
spatial_macrophage@meta.data[spatial_macrophage@meta.data$seurat_clusters == 2,'celltype'] <- 'Macro-Il31ra'
spatial_macrophage@meta.data[spatial_macrophage@meta.data$seurat_clusters == 3,'celltype'] <- 'Macro-Ccl24'
```

```r
top10$cluster <- factor(top10$cluster,levels = c('Macro_Apoe','Macro_Arg1','Macro_Il31ra'))
cluster.averages <- AggregateExpression(spatial_macrophage,features = top10$gene,
                                      group.by = 'celltype',slot = 'data',return.seurat=TRUE) 
Idents(cluster.averages) <- colnames(cluster.averages)
```
