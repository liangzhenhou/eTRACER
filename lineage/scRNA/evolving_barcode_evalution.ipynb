# evolving barcode cell frequency distribution

```python
set(allele_table['Sample'])
```

```python
cell_mutation_matrix = pd.DataFrame(index=list(set(allele_table['cellBC'])))
for gene in gene_names:
    print(gene)
    allele_table_gene = allele_table[allele_table['intBC']==gene]
    allele_table_gene = allele_table_gene[allele_table_gene['r1']!='None']
    df = allele_table_gene.pivot(index='cellBC', columns='r1', values='UMI')
    df= df.fillna(0)
    df[df > 0] = 1
    df.columns = [gene + '_' + indel for indel in df.columns]
    cell_mutation_matrix = cell_mutation_matrix.join(df, how='outer')
    cell_mutation_matrix = cell_mutation_matrix.fillna(0)
```

```python
cell_mutation_matrix
```

```python
cell_mutation_matrix = cell_mutation_matrix.astype(int)
```

```python
cell_mutation_sequence = cell_mutation_matrix.apply(lambda row: ''.join(row.astype(str)), axis=1)
cell_mutation_sequence
```

```python
evolving_barcode_count = cell_mutation_sequence.value_counts().reset_index(name='num')
cell_freq_count = evolving_barcode_count['num'].value_counts().reset_index()
cell_freq_count['count'] = cell_freq_count['num'] * cell_freq_count['count']
cell_freq_count.loc[cell_freq_count['num']>4,'num'] = 5
cell_freq_count = cell_freq_count.groupby(['num'])['count'].sum().reset_index()
cell_freq_count['freq'] = cell_freq_count['count'] / sum(cell_freq_count['count'])
cell_freq_count
```

```python
import seaborn as sns
```

```python
plt.rcParams['pdf.fonttype'] = 42
plt.gcf().set_size_inches(6, 6)
plt.figure(figsize=(4, 4))
sns.barplot(data=cell_freq_count,x='num',y='freq')
plt.ylabel('Proportion')
plt.xlabel('Number of cells with unique evolving barcode')
plt.xticks(ticks=[0,1, 2, 3, 4], labels=['1', '2', '3', '4', '>=5'])
plt.savefig('/syn1/liangzhen/jinhua_jilab_project/result/Figures/Figure2.sup/unique_evolving_barcode_cell_proportion.pdf')
plt.show()
```

# ten trees select and downsampling 200

```python
samples = ['T1_a3026','T1_a3028','T1_a3030','T2_a3026_1','T2_a3026_2','T3_a3026_1_1','T3_a3026_2_1','T2_a3028_1','T2_a3028_2','T3_a3028_1_1','T3_a3028_2_1','T2_a3030_1','T2_a3030_2','T3_a3030_1_1','T3_a3030_2_1']
```

```python
import random
```

```python
allele_table_rep = allele_table[allele_table['Sample'].isin(samples)]
cell_mutation_matrix = pd.DataFrame(index=list(set(allele_table_rep['cellBC'])))
for gene in gene_names:
    print(gene)
    allele_table_gene = allele_table_rep[allele_table_rep['intBC']==gene]
    allele_table_gene = allele_table_gene[allele_table_gene['r1']!='None']
    df = allele_table_gene.pivot(index='cellBC', columns='r1', values='UMI')
    df= df.fillna(0)
    df[df > 0] = 1
    df.columns = [gene + '_' + indel for indel in df.columns]
    cell_mutation_matrix = cell_mutation_matrix.join(df, how='outer')
    cell_mutation_matrix = cell_mutation_matrix.fillna(0)

cell_lineage = pd.read_csv('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/cell_lineage.T1_T2_T3a.csv',header=0)
cell_lineage.index = cell_lineage.cellBC
cell_lineage = cell_lineage.loc[allele_table_rep.cellBC.unique(),:]

cells_selected = []
for lineage in cell_lineage['lineageGrp'].value_counts().index[0:10]:
    cells_selected= cells_selected+ random.sample(cell_lineage[cell_lineage['lineageGrp'].isin([lineage])]['cellBC'].to_list(),200) 

cell_mutation_matrix_selected = cell_mutation_matrix.loc[cells_selected,:]
cell_mutation_matrix_selected = cell_mutation_matrix_selected.astype(int)
cell_mutation_matrix_selected = cell_mutation_matrix_selected.loc[:, (cell_mutation_matrix_selected != 0).any(axis=0)]
cell_mutation_matrix_selected.loc['synthetic'] = [ 0 for i in range(0,cell_mutation_matrix_selected.shape[1])]

cell_mutation_matrix_selected.to_csv('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/trees/clones_tree_merged/10_trees.csv')

cell_mutation_sequence = cell_mutation_matrix_selected.apply(lambda row: ''.join(row.astype(str)), axis=1)
output_file = '/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/trees/clones_tree_merged/10_trees.phy'
with open(output_file, 'w') as file:
    file.write(f"{cell_mutation_matrix_selected.shape[0]} \t{cell_mutation_matrix_selected.shape[1]}\n") 
    for index, line in cell_mutation_sequence.items():
        file.write(f"{index}\t{line}\n")  
```
