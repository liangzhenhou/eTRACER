# 8x

```python
R1 = gzip.open('/syn1/liangzhen/jinhua_jilab_project/data/DNA_Amplicon/plasmid/8x_library_SB/SQ23004121-8x-library-8x-library_combined_R1.fastq.gz')
```

```python
counter = 0
start_time = time.time()
staticBarcode_readsNum = {}
while True:
    counter += 1
    if counter % 1000000 == 0: print(': Processed {} reads in {} seconds'.format(counter, time.time()-start_time))
    r1_line = R1.readline().decode('utf-8')    
    if r1_line == '' : break
    if 'ATCTGACAGTCACTGTGACG' in r1_line:
        r1_line = r1_line[0:90]
        if 'AAAAGAAA' in r1_line and 'CACTTTTT' in r1_line:
            barcode = r1_line.split('AAAAGAAA')[0].split('ACTTTTT')[1]
            if barcode in staticBarcode_readsNum:
                staticBarcode_readsNum[barcode] += 1
            else:
                staticBarcode_readsNum[barcode] = 1
```

```python
len(staticBarcode_readsNum)
```

```python
staticBarcode_readsNum_filtered = {}
for barcode in staticBarcode_readsNum:
    if staticBarcode_readsNum[barcode] > 2:
        staticBarcode_readsNum_filtered[barcode] = staticBarcode_readsNum[barcode]
```

```python
len(staticBarcode_readsNum_filtered)
```

```python
barcode_8x = staticBarcode_readsNum_filtered.copy()
```

```python
len(barcode_8x)
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

sorted_data = np.sort(list(barcode_8x.values()))  # 从小到大排序
cumulative = np.cumsum(sorted_data)
x_axis = np.arange(1, len(sorted_data)+1)
plt.plot(cumulative,x_axis,  label="Cumulative Sum")
plt.xlabel("Cumulative reads")
plt.ylabel("barcode number")
plt.title("barcode 8x")
plt.rcParams['pdf.fonttype'] = 42
plt.gcf().set_size_inches(4, 4)
ax = plt.gca()  # 获取当前坐标轴
ax.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))  # 科学计数法
ax.ticklabel_format(style="sci", axis="y", scilimits=(0, 0))  # 强制科学计数法
plt.grid()
plt.savefig('/syn1/liangzhen/jinhua_jilab_project/result/Figures/Figure1/FigureS1_SB_8x.pdf')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

all_counts = list(barcode_8x.values()) + list(barcode_cas9.values())

min_val = max(5, min(all_counts))
max_val = 1000

start_exp = int(np.floor(np.log10(min_val)))
end_exp = int(np.ceil(np.log10(max_val)))

bins = np.logspace(start_exp, end_exp, num=(end_exp - start_exp) * 10 + 1)

cut_counts = pd.cut(all_counts, bins=bins, right=False)
count_per_bin = cut_counts.value_counts().sort_index()

x = [interval.left for interval in count_per_bin.index]
y = count_per_bin.values

plt.figure(figsize=(4, 4))
plt.scatter(x, y)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Staticbarcode reads count')
plt.ylabel('Frequency')
plt.ylim(0,100000)
average_reads_num = sum(list(barcode_8x.values())+list(barcode_cas9.values()))/(len(list(barcode_8x.values()))+len(list(barcode_cas9.values())))
plt.axvline(x=average_reads_num, color='green', linestyle='--', linewidth=2, label=f'x = {average_reads_num:.2f}')
plt.grid(True, which="both", ls="-", lw=0.5)
plt.tight_layout()
plt.savefig('/syn1/liangzhen/jinhua_jilab_project/result/Figures/Figure1/Figure1_SBbarcode_reads_stat.pdf')
plt.show()
```

```python
def get_distances(barcodes):
    n = len(barcodes)
    distance_matrix = np.zeros((n, n))

    for i in range(n):
        for j in range(i+1, n):
            dist = Levenshtein.distance(barcodes[i], barcodes[j])
            distance_matrix[i, j] = dist
            distance_matrix[j, i] = dist  # 矩阵是对称的

    # 取上三角的非对角线部分，得到所有 pairwise 距离
    pairwise_distances = distance_matrix[np.triu_indices(n, k=1)]
    return pairwise_distances
```

```python
df_long_all = pd.DataFrame()
for i in range(0,100):
    random.seed(i)
    df = pd.DataFrame()
    df['SB_8x + SB_cas9'] = list(get_distances(random.sample(list(barcode_cas9.keys())+list(barcode_8x.keys()),1000)))
    df_long = df.melt(var_name="Group", value_name="Distance")
    df_long_all = pd.concat([df_long_all,df_long],axis=0)
    
```

```python
import matplotlib.pyplot as plt
import pandas as pd

counts = df_long_all['Distance'].value_counts().sort_index()
total = counts.sum()
x = counts.index
y = counts.values / total  

bar_width = 0.8 

plt.bar(x, y, width=bar_width, color='skyblue', edgecolor='black', align='center')

plt.xlabel("Pairwise SB distance")
plt.ylabel("Ratio")
plt.xticks(x)  
plt.tight_layout()
plt.gcf().set_size_inches(5, 5)
plt.xlim(0,14.5)
plt.savefig('/syn1/liangzhen/jinhua_jilab_project/result/Figures/Figure1/Figure1_SB_distance.pdf')
plt.show()
```
