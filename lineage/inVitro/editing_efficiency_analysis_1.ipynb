```r
cut_site <- c(99,87,95,113,67,70,86,77,111,107,82,76,53,82,50)
genes <- c('Calr', 'Myl6', 'Calm2', 'Rpl39', 'Cfl1', 'Acta1', 'Hsp90b1', 'Calm1', 'Hint1', 'Btf3', 'Lgals1', 'Slc25a3', 'Tpt1', 'Hsp90ab1', 'Fth1')
cut_sites <- data.frame(row.names = genes,cut_site=cut_site)
```

```r
time <- c('T1_in_vitro')
counter <- 1
plots <- list()
for (time_point in time){
    cell_lines <- list.files(paste0('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/',time_point))
    cell_lines <- c('EPC','3T3')
    for (cell_line in cell_lines){
        indexs <- list.files(paste0('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/',time_point,'/',cell_line))
        for (index in indexs){
            genes <- list.files(paste0('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/',time_point,'/',cell_line,'/',index,'/','crispresso2'))
            genes <- genes[!grepl('html',genes)]
            genes <- grep('r1',genes,value = T)
            if (length(grep('Dox-free',genes))==0){
                genes <- genes
            }else{
                genes <- genes[-grep('Dox-free',genes)]
            }
            for (gene in genes){
                length_gene <- c()
                    Modification_count_vectors <-read.table(paste0('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/',time_point,'/',cell_line,'/',index,'/','crispresso2','/',gene,'/Modification_count_vectors.txt'),
                                                           sep='\t',header=F,row.names=1)
                    
                    length_gene <- c(length_gene,ncol(Modification_count_vectors))
                    gene_name <- unlist(strsplit(gsub("-|_","@",unlist(strsplit(gene,split = '_'))[3]),split='@'))[1]
                    Dox <- grep('Dox',unlist(strsplit(gene,split = '_')),value = T)
                    read <- grep('r1|r2',unlist(strsplit(gene,split = '_')),value = T)
                    print(as.numeric(cut_sites[gene_name,]))
                    options(repr.plot.width=5, repr.plot.height=3)
                    plot <- ggplot(as.data.frame(t(Modification_count_vectors)))+
                        geom_line(aes(x=1:length(Sequence),y=as.numeric(Deletions)/as.numeric(Total),color='Deletions'),size=1)+
                        geom_line(aes(x=1:length(Sequence),y=as.numeric(Insertions)/as.numeric(Total),color='Insertions'),size=1)+   
                        #geom_line(aes(x=1:length(Sequence),y=as.numeric(Substitutions)/as.numeric(Total),color='Substitutions'),size=1)+
                        scale_color_manual(values = c(Deletions = '#e41a1c',Insertions = '#377eb8')) +
                        geom_vline(xintercept=as.numeric(cut_sites[gene_name,]),size=1,linetype='dashed')+
                        #geom_vline(xintercept=as.numeric(cut_sites[gene_name,])-30,size=1,linetype='dashed',color='gray')+
                        #geom_vline(xintercept=as.numeric(cut_sites[gene_name,])+30,size=1,linetype='dashed',color='gray')+
                        ylim(0,1)+
                        xlim(as.numeric(cut_sites[gene_name,])-30,as.numeric(cut_sites[gene_name,])+30)+
                        labs(x='Bases position',y='Reads ratio',color='Mutation')+
                        ggtitle(paste0(time_point,'.',cell_line,'.',Dox,'.',paste0(gene_name)))+
                        theme_bw()+
                        theme(axis.text = element_text(size = 15),axis.title = element_text(size = 15),
                             legend.text = element_text(size = 15),legend.title = element_text(size = 15))
                    print(plot)
                    plots[[paste0(cell_line,'_',gene)]] <- plot
                    counter <- counter + 1
            }
        }
     
    }
   
}

```

```r
time <- c('T1_in_vitro')

quantification_all <- c()
for (time_point in time){
    cell_lines <- list.files(paste0('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/',time_point))
    for (cell_line in cell_lines){
        if (cell_line != 'EPC'){
            next
        }
        indexs <- list.files(paste0('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/',time_point,'/',cell_line))
        #print(indexs)
        for (index in indexs){
            genes <- list.files(paste0('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/',time_point,'/',cell_line,'/',index,'/','crispresso2'))
            genes <- grep('r1',genes,value = T)
            genes <- genes[!grepl('html',genes)]
            for (gene in genes){

                quantification <-read.table(paste0('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/',time_point,'/',cell_line,'/',index,'/','crispresso2','/',gene,'/CRISPResso_quantification_of_editing_frequency.txt'),
                                                           sep='\t',header=T,check.names=F)
                
                quantification$gene <- gene
                quantification$index <- index
                quantification_all <- rbind(quantification_all,quantification)
            

            }
        }
     
    }
   
}
```

```r
quantification_all$gene_name <- as.character(lapply(strsplit(quantification_all$gene,split='_'),function(x){
    return(x[3])
}))
quantification_all$gene_name <- gsub('-1','',quantification_all$gene_name)
quantification_all$Dox <- as.character(lapply(strsplit(quantification_all$gene,split='_'),function(x){
    if (length(x) == 5){
      return(x[4])  
    }else {
        return(x[5])
    }
    
}))
```

```r
genes <- c('Calr', 'Myl6', 'Calm2', 'Rpl39', 'Cfl1', 'Acta1', 'Hsp90b1', 'Calm1', 'Hint1', 'Btf3', 'Lgals1', 'Slc25a3', 'Tpt1', 'Hsp90ab1', 'Fth1')
```

```r
quantification_all$Dox <- factor(quantification_all$Dox,levels = c('Dox','Dox-free'),ordered = T)
quantification_all$gene_name <- factor(quantification_all$gene_name,levels = genes,ordered = T)
quantification_all <- quantification_all[order(quantification_all$Dox,quantification_all$gene_name),]

quantification_all$Name <- paste0(quantification_all$gene_name,'_',quantification_all$Dox)
quantification_all$Name <- factor(quantification_all$Name,levels = quantification_all$Name)
```

### 3T3

```r
time <- c('T1_in_vitro')

quantification_all <- c()
for (time_point in time){
    cell_lines <- list.files(paste0('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/',time_point))
    for (cell_line in cell_lines){
        if (cell_line != '3T3'){
            next
        }
        indexs <- list.files(paste0('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/',time_point,'/',cell_line))
        #print(indexs)
        for (index in indexs){
            genes <- list.files(paste0('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/',time_point,'/',cell_line,'/',index,'/','crispresso2'))
            genes <- grep('r1',genes,value = T)
            genes <- genes[!grepl('html',genes)]
            for (gene in genes){

                quantification <-read.table(paste0('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/',time_point,'/',cell_line,'/',index,'/','crispresso2','/',gene,'/CRISPResso_quantification_of_editing_frequency.txt'),
                                                           sep='\t',header=T,check.names=F)
                
                quantification$gene <- gene
                quantification$index <- index
                quantification_all <- rbind(quantification_all,quantification)
            

            }
        }
     
    }
   
}
```

```r
quantification_all$gene_name <- as.character(lapply(strsplit(quantification_all$gene,split='_'),function(x){
    return(x[3])
}))
quantification_all$gene_name <- gsub('-1','',quantification_all$gene_name)
quantification_all$Dox <- as.character(lapply(strsplit(quantification_all$gene,split='_'),function(x){
    if (length(x) == 5){
      return(x[4])  
    }else {
        return(x[5])
    }
    
}))
```

```r
genes <- c('Calr', 'Myl6', 'Calm2', 'Rpl39', 'Cfl1', 'Acta1', 'Hsp90b1', 'Calm1', 'Hint1', 'Btf3', 'Lgals1', 'Slc25a3', 'Tpt1', 'Hsp90ab1', 'Fth1')
```

```r
quantification_all$Dox <- factor(quantification_all$Dox,levels = c('Dox','Dox-free'),ordered = T)
quantification_all$gene_name <- factor(quantification_all$gene_name,levels = genes,ordered = T)
quantification_all <- quantification_all[order(quantification_all$Dox,quantification_all$gene_name),]

quantification_all$Name <- paste0(quantification_all$gene_name,'_',quantification_all$Dox)
quantification_all$Name <- factor(quantification_all$Name,levels = quantification_all$Name)
```

```r
quantification_all$INDEL <- quantification_all$Insertions + quantification_all$Deletions + quantification_all$`Insertions and Deletions`
quantification_all$INDEL_ratio <- quantification_all$INDEL/quantification_all$Reads_aligned *100
```
