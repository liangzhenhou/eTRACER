```python
metadata = pd.read_csv('/syn1/liangzhen/jinhua_jilab_project/result/scRNA/cellranger/metadata.T1_T2_T3a.harmony.final.refine.csv')
metadata.index = metadata.cellName.to_list()
metadata
```

```python
adata = adata[adata.obs['cellname'].isin(metadata.cellName),]
```

```python
adata.obs['state_info'] = metadata.loc[adata.obs['cellname'],'seurat_clusters_rename'].to_list()
adata.obs['state_info'] = [ 'cluster'+'_'+str(state) for state in adata.obs['state_info'] ]
adata.obs['umapharmony_1'] = metadata.loc[adata.obs['cellname'],'umapharmony_1'].to_list()
adata.obs['umapharmony_2'] = metadata.loc[adata.obs['cellname'],'umapharmony_2'].to_list()
```

```python
cell_lineage = pd.read_csv('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/cell_lineage.T1_T2_T3a.csv')
cell_lineage.index = cell_lineage['cellBC'].to_list()
cell_lineage
```

```python
adata = adata[adata.obs['cellname'].isin(cell_lineage.cellBC),]
```

```python
adata
```

```python
adata.obs['lineageGrp'] = cell_lineage.loc[adata.obs['cellname'],'lineageGrp'].to_list()
```

```python
adata.obs['time_info'] = [ sample.split('_')[0] for sample in adata.obs['Sample'] ]
```

```python
adata_sub = adata
```

```python
adata_sub = adata_sub[adata_sub.obs['time_info']!='T1',]
```

```python
adata_sub = adata_sub[adata_sub.obs['lineageGrp'].isin([1,2,3,7,5,4,6,10,9,8,19,15,13,16,28,14,17,11,40,31,26,32,20,44,29,50,23,27,30,33,35,48,49,97,37,45,62,73,55,98,109]),]
```

```python
adata_sub.obs
```

```python
adata_sub.obs['index_num'] = range(0,adata_sub.obs.shape[0])
```

```python
from scipy.sparse import coo_matrix, csr_matrix
n_cells = adata_sub.obs['index_num'].max() + 1  # 确保索引从 0 开始
n_clones = adata_sub.obs['lineageGrp'].max() + 1

# 构造稀疏矩阵
sparse_matrix = coo_matrix(
    ( [True] * len(adata_sub.obs), (adata_sub.obs['index_num'], adata_sub.obs['lineageGrp']) ),  # 数据和值 (全为 True)
    shape=(n_cells, n_clones)  # 指定矩阵形状
)
adata_sub.obsm["X_clone"] = csr_matrix(sparse_matrix)
```

```python
adata_sub.uns['data_des'] = 'test'
```

```python
adata_sub.obs
```

```python
selected_times = "T2"
selected_fates = [
    "cluster_1",
    "cluster_2",
    "cluster_3",
    "cluster_4",
    "cluster_5",
]
cs.tl.fate_coupling(
    adata_sub,
    source="X_clone",
    selected_fates=selected_fates,
    selected_times=selected_times,
    normalize=False,
)
cs.pl.fate_coupling(adata_sub, source="X_clone")
```

```python
import seaborn as sns
from matplotlib.colors import ListedColormap,LinearSegmentedColormap
```

```python
df = pd.DataFrame(adata_sub.uns['fate_coupling_X_clone']['X_coupling'])
df.columns = ['cluster_1', 'cluster_2', 'cluster_3', 'cluster_4', 'cluster_5']
df.index = ['cluster_1', 'cluster_2', 'cluster_3', 'cluster_4', 'cluster_5']
df = df.loc[['cluster_4','cluster_2', 'cluster_1','cluster_3','cluster_5'],['cluster_4','cluster_2', 'cluster_1','cluster_3','cluster_5']]
df.columns = ['state_4','state_2', 'state_1','state_3','state_5']
df.index = ['state_4','state_2', 'state_1','state_3','state_5']

colors = [
    (0.0, "black"),      # 0.0 到 0.9 为黑色
    (0.1, "gray"),
    (0.95, "white"),      # 0.91 开始渐变色
    (1.0, "#de2d26")         # 1.0 为红色
]



cmap = LinearSegmentedColormap.from_list("custom_cmap", colors, N=256)

plt.rcParams['pdf.fonttype'] = 42
plt.gcf().set_size_inches(5, 5)
# 创建 clustermap
g = sns.clustermap(
    df,col_cluster=False,row_cluster=False,
    cmap=cmap,          # 使用自定义颜色映射
    vmin=0,             # 最小值
    vmax=1,             # 最大值
    #annot=True,         # 在热图中显示数值
    fmt=".3f",         # 数值显示格式（保留 3 位小数）
    figsize=(5, 5)      # 图形大小
)
ticks = [0.0, 0.1, 0.9, 0.95, 1.0]
plt.savefig('/syn1/liangzhen/jinhua_jilab_project/result/Figures/Figure3/Figure3_to_jinhua/clonal_coupling_T2.pdf')

# 显示图形
plt.show()
```
