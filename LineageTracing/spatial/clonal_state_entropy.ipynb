```r
freq_mat <- T2_spatial_cell_lineage %>%
  group_by(lineageGrp, state) %>%
  summarise(count = n(), .groups = "drop") %>%
  group_by(lineageGrp) %>%
  mutate(prop = count / sum(count)) %>% as.data.frame
freq_mat <- freq_mat[,c(1,2,4)]
freq_mat <- as.data.frame(pivot_wider(data = freq_mat, id_cols = lineageGrp,names_from = state, values_from = prop, values_fill = 0))
rownames(freq_mat) <- freq_mat$lineageGrp
freq_mat <- freq_mat[,c(2,3,4,5,6)]
freq_mat
```

```r
library(entropy)

entropy_vec <- as.data.frame(apply(freq_mat, 1, function(p) entropy.empirical(p, unit = "log2")))
colnames(entropy_vec) <- 'state_composition_entropy'
entropy_vec                                  
```

```r
cell_distance_to_bundary <- read.table('/syn1/liangzhen/jinhua_jilab_project/result/DNA_Amplicon/SPATIAL/T2/cell_distance_to_bundary.csv',sep=',',header=T,row.names=1)
rownames(cell_distance_to_bundary) <- cell_distance_to_bundary$x_y
T2_spatial_cell_lineage$cell_distance_to_bundary <- cell_distance_to_bundary[rownames(T2_spatial_cell_lineage),'distances']
T2_spatial_cell_lineage
```

```r
library(dplyr)

clone_distance_var <- T2_spatial_cell_lineage %>%
  group_by(lineageGrp) %>%
  summarise(variance = var(cell_distance_to_bundary))
clone_distance_var <- as.data.frame(clone_distance_var)
clone_distance_var
```

```r
clone_distance_var$state_composition_entropy <- entropy_vec[clone_distance_var$lineageGrp,'state_composition_entropy']
clone_distance_var <- clone_distance_var[!clone_distance_var$lineageGrp %in% c('EE_a3026_C30','EE_a3030_C42'),]
```
