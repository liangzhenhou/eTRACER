```r
freq_mat <- T2_spatial_cell_lineage %>%
  group_by(lineageGrp, state) %>%
  summarise(count = n(), .groups = "drop") %>%
  group_by(lineageGrp) %>%
  mutate(prop = count / sum(count)) %>% as.data.frame
freq_mat <- freq_mat[,c(1,2,4)]
freq_mat <- as.data.frame(pivot_wider(data = freq_mat, id_cols = lineageGrp,names_from = state, values_from = prop, values_fill = 0))
rownames(freq_mat) <- freq_mat$lineageGrp
freq_mat <- freq_mat[,c(2,3,4,5,6)]
freq_mat
```

```r
jsd_matrix <- distance(freq_mat, method = "jensen-shannon")
rownames(jsd_matrix) <- colnames(jsd_matrix) <- rownames(freq_mat)
jsd_matrix
```

```r
mat <- as.matrix(jsd_matrix)

# 获取上三角的索引
upper_idx <- which(upper.tri(mat), arr.ind = TRUE)

# 构建长格式 data.frame
long_df <- data.frame(
  clone1 = rownames(mat)[upper_idx[, 1]],
  clone2 = colnames(mat)[upper_idx[, 2]],
  value  = mat[upper.tri(mat)]
)
long_df['sample1'] = as.character(as.data.frame(strsplit(long_df[,'clone1'],split = '_'))[2,])
long_df['sample2'] = as.character(as.data.frame(strsplit(long_df[,'clone2'],split = '_'))[2,])
long_df = long_df[long_df['sample1']==long_df['sample2'],]
print(long_df)
```
